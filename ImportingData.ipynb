```python
import pandas as pd
import numpy as np
```

```python
file_path = ("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")
```

```python
df = pd.read_csv(file_path, header=None)
```

```python
print("The first five rows of the dataframe")
df.head(5)
```

```python
df.tail(10)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.columns
```

```python
df.head(10)
```

```python
df1=df.replace('?',np.NaN)
```

we can drop missing values along the column "price" as follows:


```python
df = df1.dropna(subset=["price"], axis=0)
df.head(20)
```

here axis=0 means that the contents along the entire row will be dropped wherever the entity 'price' is found to be NaN.

```python
print(df.columns)
```

```python
df.to_csv("automobile.csv",index=False)
```
